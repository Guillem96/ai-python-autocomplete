```python
from pathlib import Path
import tqdm.auto as tqdm
```

```python
input_path = Path('data/preprocessed')
chars = set()
```

```python
files = list(input_path.iterdir())
for fname in tqdm.tqdm(files):
    with fname.open() as f:
        fchars = list(f.read())
        chars.update(fchars)
```

```python
print(f'# Unique chars {len(chars)}')
```
